```python
room_images = {
    "game_room" : "Image_game_room.jpg",
    "bedroom_1" : "Image_bedroom_1.png",
    "bedroom_2" : "Image_bedroom_2.jpg",
    "living_room" : "Image_living_room.png"
}

#Function to show an image:
def show_room_image(room_name):
    from IPython.display import Image
    
    path = "..\Img\\" + room_images[room_name]
    display(Image(filename = path, width=300, height=300))
```

```python
show_room_image("bedroom_1")
show_room_image("bedroom_2")
```

```python
show_room_image("game_room")
show_room_image("living_room")
```
